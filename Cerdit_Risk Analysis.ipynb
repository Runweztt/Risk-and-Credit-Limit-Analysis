```python
# Add payment behavior metrics
df['AT_RISK'] = (df['CREDIT_UTILIZATION'] > high_utilization_threshold) | \
                (df['PRC_FULL_PAYMENT'] < 0.70)  # Assuming 70% is a threshold for full payments

# Get a summary of high-risk customers
risk_summary = df[df['AT_RISK']][['CUST_ID', 'BALANCE', 'CREDIT_LIMIT', 'CREDIT_UTILIZATION', 'PAYMENTS', 'MINIMUM_PAYMENTS']]
print("\nHigh-Risk Customers based on Credit Utilization and Payment Behavior:")
print(risk_summary)
```

```python
# Define risk categories based on Credit Utilization and Payments
def categorize_risk(row):
    if row['CREDIT_UTILIZATION'] > 0.8 and row['PAYMENTS'] < row['MINIMUM_PAYMENTS']:
        return 'High Risk'
    elif row['CREDIT_UTILIZATION'] > 0.6:
        return 'Moderate Risk'
    else:
        return 'Low Risk'

df['RISK_CATEGORY'] = df.apply(categorize_risk, axis=1)

# Display counts of each risk category
risk_counts = df['RISK_CATEGORY'].value_counts()
print(risk_counts)
```
